```python
import pandas_datareader.data as web
import pandas as pd
```

```python
all_data={ticker: web.get_data_yahoo(ticker)
         for ticker in ["AAPL","IBM","MSFT","GOOG"]}
price = pd.DataFrame({ticker : data['Adj Close']
                                    for ticker, data in all_data.items()})
volume = pd.DataFrame({ticker : data['Volume']
                                     for ticker, data in all_data.items()})
```

```python
price.head(),volume.head()
```

```python
returns = price.pct_change()  #주식의 퍼센트 변화율 계산 함수
returns.head()
```

```python
url = "https://en.wikipedia.org/w/index.php?title=List_of_Game_of_Thrones_episodes&oldid=802553687"
r2 = requests.get(url)
r2.text
html_soup = BeautifulSoup(r2.text,'html.parser')
found_h1 = html_soup.find("h1") # h1태그 줄을 찾아라
found_h1
```

```python
found_h1.contents # h1태그의 컨텐츠 내용
```

```python
for found in html_soup.find_all('h2'):
    print(found.text)     #h2에 있는 것들을 꺼냄
```

```python
episodes = []
ep_tables = html_soup.find_all('table', 
                             class_ = "wikitable plainrowheaders wikiepisodetable")
ep_tables
```

```python
for table in ep_tables:
    headers = []
    rows = table.find_all('tr')
    for header in table.find('tr').find_all('th'):
        headers.append(header.text)
        
    for row in table.find_all('tr')[1:]:
        values = []
        for col in row.find_all(['th','td']):
            values.append(col.text)
        if values:
            episode_dict = {headers[i]:values[i] for i in range(len(values))}
            episodes.append(episode_dict)
            
for episode in episodes:
    print(episode)
```

```python
import pandas as pd
df = pd.DataFrame(episodes)
df.head()
```
