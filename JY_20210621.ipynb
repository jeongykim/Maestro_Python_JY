```python
# 106pg
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, 
    random_state=42)
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
print(tree.score(X_train, y_train))
print(tree.score(X_test, y_test))
```

```python
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)
print(tree.score(X_train, y_train))
print(tree.score(X_test, y_test))
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file='tree.dot', class_names=['악성', '양성'],
               feature_names=cancer.feature_names, impurity=False, filled=True)
```

```python
import os 
import pandas as pd
ram_prices = pd.read_csv('ram_price.csv')
plt.semilogy(ram_prices.date, ram_prices.price)
plt.show()
```

```python
plt.plot(ram_prices.date, ram_prices.price)
```

```python
np.log(ram_prices.price)
```

```python
from sklearn.tree import DecisionTreeRegressor
data_train = ram_prices[ram_prices.date<2000]
data_test = ram_prices[ram_prices.date>=2000]
```

```python
data_train.shape
```

```python
data_test.shape
```

```python
X_train = data_train.date.to_numpy()
X_train
```

```python
# 2차원 배열로 바꿔줌
X_train = X_train[:,np.newaxis]
```

```python
# 데이터와 타깃 사이 관계를 간단하게 해주기 위해 log스케일 변환
y_train = np.log(data_train.price)
```

```python
from sklearn.linear_model import LinearRegression
dt = DecisionTreeRegressor().fit(X_train,y_train)
lr = LinearRegression().fit(X_train,y_train)
```

```python
#모든 X의 데이터 
X_all = ram_prices.date.to_numpy()[:,np.newaxis]
```

```python
pred_dt = dt.predict(X_all)
pred_lr = lr.predict(X_all)
```

```python
# 로그 스케일 취한 결과르 다시 지수함수를 통해 복구
price_dt = np.exp(pred_dt)
price_lr = np.exp(pred_lr)
```

```python
# 이를 보면 트리모델은 시계열 데이터를 가지고 예측을 할 때는 잘 맞지 않다는 것을 보여준다
plt.semilogy(data_train.date,data_train.price, label = "train data")
plt.semilogy(data_test.date,data_test.price, label = "test data")
plt.semilogy(X_all,price_dt, label = "pred of dt")
plt.semilogy(X_all,price_lr, label = "pred of lr")
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons
X,y = make_moons(n_samples=100,noise=0.25,random_state=5)
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y, random_state=42)
forest = RandomForestClassifier(n_estimators=5,random_state=2).fit(X_train,y_train)
fig, axes = plt.subplots(2,3,figsize=(20,10))
for i, (ax, tree) in enumerate(zip(axes.ravel(),forest.estimators_)):
    mglearn.plots.plot_tree_partition(X,y,tree,ax=ax)
mglearn.plots.plot_2d_separator(forest,X,fill=True)
mglearn.discrete_scatter(X[:,0],X[:,1],y)
plt.show()
```
