```python
dataA = np.random.standard_normal((100,2))
dataB = np.random.standard_normal((100,2))
dataC = np.random.standard_normal((100,2))
dataA += np.array((-3,-3))
dataB += np.array((4,4))
dataC += np.array((-1,4))
plt.scatter(dataA[:,0],dataA[:,1], color="b")
plt.scatter(dataB[:,0],dataB[:,1], color="0.7") # color에 숫자 값을 넣으면 검은색에서 투명도 지정이 됨 0~1 
plt.scatter(dataC[:,0],dataC[:,1], color="#387966") # 원하는 색상표를 넣을 수도 있음
plt.axhline(1, color="#BC8F8F")
plt.axvline(1.5, color="#79AAFF")
plt.show()
```

```python
import matplotlib.cm as cm
N=256
angle = np.linspace(0,8*2*np.pi,N)
radius = np.linspace(5,1,N)
X = radius*np.cos(angle)
Y = radius*np.sin(angle)
plt.scatter(X,Y, c= angle, cmap=cm.Spectral) #cm.Greys,cm.Spectral,cm.Greys
plt.show()
```

```python
import matplotlib.patches as patches
plt.axes(polar= True)

radius = [1, 0.3, 0.2, 0.6, 0.1, 0.2, 0.6, 0.8, 0.5]
theta = np.linspace(0, 2*np.pi, len(radius), endpoint= False)
points = np.vstack((theta, radius)).T # 행렬 차원 맞춘다. 일단 pass
plt.gca().add_patch(patches.Polygon(points, color= '0.5'))
plt.show()
```
