```python
num = int(input("몇 개?"))
for hang in range(1, num+1):
    #숫자 출력
    for yol in range(1, hang+1):
        print(yol,end=" ")
    #별 출력
    for yol in range(1,num+1-hang):
        print("*",end=" ")
    print("")
```

```python
def add_many(mode,*a): #*a는 하나만 가져오는 것이 아니라, 덩어리로 가져온다.
    if mode =="sum":
        result=0
        for item in a: result+=item
    elif mode =="mul":
        result=1
        for item in a: result*=item
    return result
print(add_many("sum",1,2))
print(add_many("mul",1,2,3,4))
```

```python
def add_many(mode,*a): # 안되는 이유는 동작 순서가 맞지 않기 때문에
    if mode =="sum":
        result=0
        for item in a: result+=item
    elif mode =="mul":
        result=1
        for item in a: result*=item
    return result
print(add_many(1,2,"sum"))
print(add_many(1,2,3,4,"mul"))
```

```python
f =  open("새파일.txt", "r")
list1 = []
list2 = []
while True:
    line1 = f.readline()
    if not line1: break
    print(line1)
    list1.append(line1)
f.close()
for item in list1:
    processed = item.replace("\n", " ")
    print(processed,end=" ")
    list2.append(processed)   
print(list2)
```
