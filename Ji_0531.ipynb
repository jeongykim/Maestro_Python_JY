```python
import matplotlib.cm as cm  # color map
N = 256
angle = np.linspace(0, 8*2*np.pi, N)
radius = np.linspace(5, 1, N)
X = radius*np.cos(angle)
Y = radius*np.sin(angle)
plt.scatter(X, Y, c= angle, cmap= cm.Spectral) # cm.Greys, cm.hsv
plt.show()
```

```python
M = np.random.standard_normal((500, 2))
R = np.sum(M**2, axis= 1)
plt.scatter(M[:, 0], M[:, 1], c= '0.7', marker= 's', s= 32*R, edgecolor= 'k')
plt.show()
```

```python
M = np.random.standard_normal((500, 2))
R = np.sum(M**2, axis= 1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.scatter(M[:, 0], M[:, 1], c= '0.7', marker= 's', s= 32*R, edgecolor= 'k')
plt.show()
```

```python
T = np.linspace(0, 2*np.pi, 1024) # 한바퀴 돌리기 위한 범위 0 : 2pi
plt.axes(polar= True) # polar 극좌표계
plt.plot(T, 1+0.25*np.sin(16*T))
plt.show()
```

```python
import matplotlib.patches as patches
plt.axes(polar= True)

radius = [1, 0.5, 0.7, 0.3, 0.1, 0.2, 0.6, 0.8, 0.5]
theta = np.linspace(0, 2*np.pi, len(radius), endpoint= False)
points = np.vstack((theta, radius)).T # 행렬 차원 맞춘다. 일단 pass
plt.gca().add_patch(patches.Polygon(points, color= '0.5'))
plt.show()
```
