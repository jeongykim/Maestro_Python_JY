```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
import mglearn
X,y= make_moons(n_samples=100,noise=0.25,random_state=3)
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,random_state=42)
mlp = MLPClassifier(solver='lbfgs',random_state=0,max_iter=3,
                   hidden_layer_sizes=[10])
#하나의 은닉층의 10개 유닛
mlp.fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp,X_train)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
```

```python
mlp = MLPClassifier(solver='lbfgs',random_state=0,
                   hidden_layer_sizes=[10,10])
#두개의 히든레이어에 10개 유닛들
mlp.fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp,X_train)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
```

```python
mlp = MLPClassifier(solver='lbfgs',random_state=0, activation='tanh',
                   hidden_layer_sizes=[10,10])
#두개의 히든레이어에 10개 유닛들
mlp.fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp,X_train)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2,5,figsize=(25,8))
for ax_y,n_nodes in zip(axes,[10,100]):
    for ax_x, alpha in zip(ax_y,[0.0001,0.01,0.1,1,10]):
        mlp = MLPClassifier(solver='lbfgs',random_state=0, alpha = alpha,
                   hidden_layer_sizes=[n_nodes,n_nodes])
        mlp.fit(X_train,y_train)
        mglearn.plots.plot_2d_separator(mlp,X_train, ax=ax_x)
        mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train, ax=ax_x)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_circles
X,y=make_circles(noise=0.25,factor=0.5,random_state=1)
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
model = GradientBoostingClassifier(random_state=0).fit(X_train,y_train)
print(X_train.shape)
print(X_test.shape)
print(model.decision_function(X_test).shape)
```

```python
# 결정함수는 양수는 1, 음수는 0으로 파악함
rint(model.decision_function(X_test))
print(y_test)
```

```python
mglearn.discrete_scatter(X[:,0],X[:,1],y)
```

```python
print(model.decision_function(X_test)>0)
print(y_test)
```

```python
fig, axes = plt.subplots(1,2,figsize=(13,5))
mglearn.tools.plot_2d_separator(model,X,ax=axes[0],alpha=0.4,
                               fill=True,cm=mglearn.cm2)
score_image = mglearn.tools.plot_2d_scores(model,X,ax=axes[1],alpha=0.4,
                                          cm=mglearn.ReBl)
for ax in axes:
    mglearn.discrete_scatter(X_test[:,0],X_test[:,1],y_test,
                        markers='^',ax=ax)
    mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train,
                        markers='o',ax=ax)
cbar=plt.colorbar(score_image,ax=axes.tolist())
```
